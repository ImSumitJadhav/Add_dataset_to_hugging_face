```python
from huggingface_hub import login
from datasets import Dataset
login()
#give hugging face token which is in "write" type
```

```python
import pandas as pd

input = pd.read_csv('/content/stocknews_summarization_data_2000.csv',)
input = input.drop('Unnamed: 0',axis = 1)
input
```

```python
dataset = Dataset.from_pandas(input)
dataset = dataset.shuffle()
dataset = dataset.train_test_split(test_size=0.15)
```

```python
dataset.push_to_hub(token="hugging_face_token",repo_id='model_repo_id') #give hugging face token which is in "write" type
```
